Faites afficher la courbe de la fonction *sinus(x)* entre *x = 0* et *x = 20*

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 20, 100)
plt.plot(x, np.sin(x))
plt.show()
```

Placer les valeurs de *sin(x)* pour les entiers de 2 à 10 dans un tableau

```python
import numpy as np
import matplotlib.pyplot as plt

# Initialisation du tableau
sinus = np.zeros(11)

# Remplir le tableau avec les valeurs de sin(i) pour i allant de 2 à 9
for i in range(2, 11):
    sinus[i] = np.sin(i)

# Afficher le tableau des valeurs de sin(x)
print("Tableau des valeurs de sin(x) pour x de 2 à 10 :", sinus)
```

```python
# Tracer les valeurs de sin(x)
x_values = np.arange(2, 11)
plt.plot(x_values, sinus[2:11], marker='o')
plt.xlabel('x')
plt.ylabel('sin(x)')
plt.title('Valeurs de sin(x) pour x de 2 à 10')
plt.grid(True)
plt.show()
```
